```python
import requests
from bs4 import BeautifulSoup
```

```python
headers = {
    "User-Agent":"Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/76.0.3809.100 Safari/537.36"
}
```

```python
class Movie:
    
    def __init__(self):
        self.rank = 0
        self.star = 0
        self.title = None
        self.stars = None
        self.time = None
    
    def __str__(self):
        return ("{}\t《{}》\t{}\t{}\t{}".format(self.rank,self.title,self.stars,self.time,self.star))
```

```python
def parse(url):
    response = requests.get(url,headers = headers)
    if response.status_code == 200:
        movie_list = []
        html = response.text
        soup = BeautifulSoup(html,'lxml')
        dl = soup.find('dl',{'class':'board-wrapper'})
        for dd in dl.find_all('dd'):
            movie = Movie()
            rank = dd.i.text.strip()
            body = dd.find('div',{'class':'movie-item-info'})
            star = dd.find('div',{'class':'movie-item-number score-num'}).text.strip()
            title,stars,time = tuple(body.find_all('p'))
            movie.rank = rank
            movie.title = title.a.get('title')
            movie.stars = stars.text.strip()
            movie.time = time.text.strip()
            movie.star = star
            movie_list.append(movie)
        return movie_list
    else:
        return None
```

```python
movie_list = []
for page_num in range(10):
    url = 'https://maoyan.com/board/4?offset={}'.format(page_num*10)
    movies = parse(url)
    if movies is not None:
        movie_list.extend(movies)
```

```python
for movie in movie_list:
    print(movie)
```
